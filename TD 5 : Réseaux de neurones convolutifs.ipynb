# Réseau de neurones

## Description du modèle

Dans cette section, nous avons créé un modèle de réseau de neurones convolutif simple (**SimpleCNN**) en utilisant PyTorch. Ce modèle est conçu pour effectuer de la classification d'images. Voici une description détaillée de l'architecture :

- **Initialisation du modèle** :
  - Le modèle commence par trois couches convolutifs successives :
    - La première convolution (`conv1`) prend une image en entrée avec 1 canal (par défaut) et produit 32 cartes de caractéristiques. La taille du noyau est de 5 avec un padding de 2.
    - La deuxième convolution (`conv2`) transforme les 32 cartes de caractéristiques en 64 avec un noyau de taille 3 et un padding de 1.
    - La troisième convolution (`conv3`) maintient 64 cartes de caractéristiques avec les mêmes paramètres.
  - Une couche fully connected finale (`final`) permet de réduire les 4096 caractéristiques en sortie à un nombre de classes définies par l'utilisateur (par défaut, 10).

- **Propagation avant (Back - Propagation)** :
  - L'image passe d'abord par la première couche de convolution avec une activation `leaky_relu`. La taille de l'image passe de **1×32×32** à **32×32×32**.
  - Une opération de pooling max (taille de noyau = 2, stride = 2) réduit la taille à **32×16×16**.
  - La deuxième couche de convolution transforme les caractéristiques à **64×16×16**, suivie d'un autre pooling max qui les réduit à **64×8×8**.
  - La troisième couche de convolution applique une dernière transformation sans changement de taille (**64×8×8**).
  - Les cartes de caractéristiques sont ensuite aplaties en un vecteur de taille **4096** avant d’être passées à la couche fully connected pour générer les prédictions finales.

```python
class SimpleCNN(nn.Module):
    def __init__(self,n_out=10, n_in=1):
        super().__init__()

        # Put the layers here
        self.conv1 = nn.Conv2d(n_in, 32, kernel_size=5, padding=2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3, padding=1)

        self.final = nn.Linear(4096, n_out)

    def forward(self, x):
        # Put the sequence of layer computation here (conv, pooling, activation, linear...)
        # Each step transforms the input as : x = f(x)

        x = F.leaky_relu(self.conv1(x)) ## l'image 1x32x32 devient 32x32x32
        x = F.max_pool2d(x, kernel_size=2, stride=2) ## puis 32x16x16
        x = F.leaky_relu(self.conv2(x)) ## puis devient 64x16x16
        x = F.max_pool2d(x, kernel_size=2, stride=2) ## puis devient 64x8x8
        x = F.leaky_relu(self.conv3(x)) ## pas de changement

        x = x.view(-1,4096) ## 64x8x8 devient 4096

        x = self.final(x) ## on finit exactement de la même façon

        return x
```

## Entraînement

### Boucle principale d'optimisation

Cette section décrit la boucle principale pour l'optimisation du modèle. Elle suit les mêmes principes que ceux utilisés pour le MLP dans le cours précédent.

#### Entraînement

1. Les gradients sont initialisés à zéro.
2. Les sorties du modèle sont calculées.
3. La perte est calculée avec la fonction d'entropie croisée.
4. La méthode `backward` est appelée sur la perte pour le calcul des gradients.
5. La méthode `step` est utilisée sur l'optimiseur pour mettre à jour les paramètres du modèle.
6. Les paramètres d'optimisation sont ajustés à l'aide du scheduler.
7. Les prédictions sont calculées à partir des sorties (sous format numpy), en prenant l'indice du maximum (`argmax`) du vecteur de prédiction.
8. La matrice de confusion est mise à jour.

#### Validation

1. Les sorties sont calculées sur le jeu de validation.
2. Les prédictions sont obtenues à partir des sorties (sous format numpy), en prenant l'indice du maximum (`argmax`) du vecteur de prédiction.
3. La matrice de confusion est mise à jour.

#### Affichage des résultats

1. Les précisions (accuracies) sur les ensembles d'entraînement et de validation sont calculées et affichées.
2. Ces précisions sont enregistrées dans les listes correspondantes.

```python
import copy
from IPython.display import clear_output

## To compute the current error from confusion matrix
def accuracy(cm):
    return cm.trace() / cm.sum()


def train_model_multiclass(model, nlabel, trainloader, valloader,
                           optimizer, scheduler, num_epochs=25, display=False):
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_perf = 0.0

    # list for saving accuracies
    train_perf = []
    test_perf = []
    train_losses = []

    criterion = nn.CrossEntropyLoss()

    # iterate over epochs
    for epoch in range(num_epochs):

        # set the model in train mode
        model.train()

        # create a zero confusion matrix
        cm = np.zeros((nlabel,nlabel))

        running_loss = 0.0

        for inputs, targets in tqdm(trainloader, ncols=80,  desc="Epoch {}".format(epoch)):

            # Put data to GPU
            inputs = inputs.to(device)
            targets = targets.to(device)

            # gradients to zero
            optimizer.zero_grad()

            # compute outputs
            outputs = model(inputs)

            # compute loss
            loss = criterion(outputs, targets)

            # backward on loss
            loss.backward()
            running_loss += loss.item() * inputs.size(0)

            # optimizer step
            optimizer.step()

            # Transfer data to main CPU and translate tensors to numpy
            outputs_np = outputs.cpu().detach().numpy()
            targets_np = targets.cpu().detach().numpy()

            # compute the predictions
            pred = np.argmax(outputs_np, axis=1)

            # update the perf
            cm = cm + confusion_matrix(targets_np, pred, labels=list(range(nlabel)))

        # set the model to evaluation mode
        model.eval()

        # create the perf
        cm_test = np.zeros((nlabel,nlabel))

        # tell not to reserve memory space for gradients (much faster)
        with torch.no_grad():
            for inputs, targets in tqdm(valloader, ncols=80, desc=" Test {}".format(epoch)):

                # Put data to GPU
                inputs = inputs.to(device)
                targets = targets.to(device)

                # compute outputs
                outputs = model(inputs)

                # Transfer data to main CPU and translate tensors to numpy
                outputs_np = outputs.cpu().detach().numpy()
                targets_np = targets.cpu().detach().numpy()

                # compute the predictions
                pred = np.argmax(outputs_np, axis=1)

                # update the confusion matrix
                cm_test = cm_test + confusion_matrix(targets_np, pred, labels=list(range(nlabel)))

        clear_output()

        # compute accuracies and display them
        oa_train = accuracy(cm)
        oa_test = accuracy(cm_test)
        epoch_loss = running_loss / len(trainloader.dataset)
        print("")
        print("train = {:.2f}%; val = {:.2f}%; loss = {:e}".format(oa_train*100, oa_test*100, epoch_loss))

        # add accuracies to the lists
        train_perf.append(oa_train)
        train_losses.append(epoch_loss)
        test_perf.append(oa_test)

        # Plot the learning curves
        if display:
          plt.plot(train_perf,label='train')
          plt.plot(test_perf,label='test')
          plt.legend()
          plt.show()

        # deep copy the model
        if oa_test > best_perf:
            best_perf = oa_test
            best_model_wts = copy.deepcopy(model.state_dict())

        # learning rate update
        scheduler.step()

    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_perf))

    # load best model weights
    model.load_state_dict(best_model_wts)

    # return model and learning curves
    return model, train_perf, test_perf, train_losses
```

## Partie Code

Le modèle a été entraîné en utilisant un réseau de neurones convolutifs **SimpleCNN** sur le jeu de données GTSRB (German Traffic Sign Recognition Benchmark). Le code ci-dessous commence par tenter de charger un modèle préalablement appris à partir d'un fichier sauvegardé à l'aide de `torch.load`. Si aucun modèle n'est trouvé, l'entraînement démarre à partir de zéro. L'optimiseur **Adam** est utilisé avec un taux d'apprentissage de `1e-3` et un scheduler `StepLR`, qui ajuste le taux d'apprentissage de moitié toutes les 5 époques pour optimiser la convergence. L'entraînement est géré par la fonction `train_model_multiclass`, qui prend en charge l'optimisation, le calcul des pertes et des erreurs, ainsi que l'affichage des résultats d'entraînement. Enfin, une évaluation du modèle est effectuée sur l'ensemble de test, et une matrice de confusion est générée pour comparer les prédictions aux vraies étiquettes.

```python
from torch.optim import lr_scheduler

# The name of the dataset
tag = "simpleCNN_gtsrb"
img_path = DATA_DIR

nlabel = 43

# create the network
net_gtsrb = SimpleCNN(n_in = 3, n_out=nlabel)

# Read the last learned network (if stored)
try:
  if (os.path.exists(os.path.join(img_path, 'network_{:s}.pth'.format(tag)))):
    print('Resume from last learning step')
    net_gtsrb = torch.load(os.path.join(img_path, 'network_{:s}.pth'.format(tag)))
except:
  print('Learn from scratch')
  net_gtsrb = SimpleCNN(n_in = 3, n_out=nlabel)

net_gtsrb.to(device)

# Define optimization components (to be used in the learning function)
optimizer = torch.optim.Adam(net_gtsrb.parameters(), lr=1e-3)
scheduler = lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5, verbose=True)

# Learning
max_epoch = 5

learned_model, train_error, test_error, train_losses = \
train_model_multiclass(net_gtsrb, nlabel,
                        trainloader, testloader,
                        optimizer, scheduler,
                       num_epochs=max_epoch, display=True)

# Save last learned model
torch.save(learned_model, os.path.join(img_path, 'network_{:s}.pth'.format(tag)))

net_gtsrb = learned_model

# Evaluation
net_gtsrb.eval()

allpred = []
gt = []
with torch.no_grad():  ### no need to compute gradients
    for inputs, targets in testloader:
        inputs = inputs.to(device)
        outputs = net_gtsrb(inputs)
        _,pred = outputs.max(1)
        allpred += pred.cpu().numpy().tolist()
        gt += targets.cpu().numpy().tolist()

cm = confusion_matrix(allpred,gt,labels = list(range(nlabel)))
print(cm)
```

# Apprentissage par transfert

Nous étudions dans cette partie la capacité d'un réseau à transférer ses connaissances à une autre tâche. Plus précisément, nous allons examiner si les caractéristiques apprises sur un ensemble de données peuvent aider à l'apprentissage sur un autre ensemble de données.

Ici, nous allons étudier le transfert des connaissances depuis GTSRB (panneaux de signalisation colorés) vers SVHN (chiffres colorés sur des maisons), en supposant que les bonnes caractéristiques extraites des panneaux de signalisation sont également utiles pour les chiffres colorés.

## Comparaison des stratégies d'Apprentissage

Dans cette section, nous comparons trois stratégies d'apprentissage pour le transfert des connaissances depuis un réseau pré-entraîné sur le jeu de données GTSRB (panneaux de signalisation) vers le jeu de données SVHN (chiffres colorés). Ces stratégies sont :
1. **Fine-tuning (FT)** : mise à jour de l'ensemble du réseau, y compris les couches convolutives et la couche de décision.
2. **Last layer (LAST)** : mise à jour uniquement de la dernière couche du réseau, en supposant que les caractéristiques extraites par les couches convolutionnelles sont adaptées à la nouvelle tâche.
3. **From scratch (SCRATCH)** : apprentissage d'un nouveau réseau à partir d'une initialisation aléatoire.

---

## Stratégies d'apprentissage

### Fine-tuning (FT)

Cette stratégie consiste à initialiser le réseau à partir du modèle pré-entraîné sur GTSRB, puis à mettre à jour toutes les couches, y compris les couches convolutionnelles et la couche de décision. La dernière couche est remplacée par une couche adaptée au nombre de classes du jeu de données SVHN (10 chiffres).

- **Avantage** : Permet une adaptation complète du réseau à la nouvelle tâche.

### Mise à jour de la dernière couche (LAST)

Seule la dernière couche est mise à jour dans cette stratégie, tandis que les couches convolutionnelles sont gelées. Cela repose sur l’hypothèse que les caractéristiques extraites sur GTSRB sont directement réutilisables pour SVHN.

- **Avantage** : Réduction significative du temps de calcul, car seules les couches de décision sont ajustées.

### Apprentissage à partir de zéro (SCRATCH)

Le réseau est initialisé aléatoirement, et toutes les couches sont apprises depuis le début. Aucun transfert de connaissance n'est utilisé dans cette stratégie.

- **Avantage** : Ne dépend pas d'un pré-entraînement, ce qui peut être utile lorsque les données des deux tâches diffèrent fortement.

---

```python
from torch.optim import lr_scheduler

tag_pretrained = "simpleCNN_gtsrb"
nlabel = 10

img_path = DATA_DIR

update = 'FT'
update = 'LAST'
update = 'SCRATCH'

for update in ['FT', 'LAST', 'SCRATCH']:
  ## Update the whole pre-trained network
  if update == 'FT':
    print(f'Fine tuning from pre-trained network {tag_pretrained}')
    tag = "simpleCNN_svhn_ft"

    # loads network learned from gtsrb
    net_svhn = torch.load(os.path.join(img_path, 'network_{:s}.pth'.format(tag_pretrained)))

    # Replace the last layer
    net_svhn.final = nn.Linear(4096, nlabel)

  ## Update only the last layer
  elif update == 'LAST':
    print(f'Learn last layer from pre-trained network {tag_pretrained}')
    tag = "simpleCNN_svhn_last"

    # loads network learned from gtsrb
    net_svhn = torch.load(os.path.join(img_path, 'network_{:s}.pth'.format(tag_pretrained)))

    # only allows last layer to be learned
    for param in net_svhn.parameters():
        param.requires_grad = False

    # Replace the last layer
    net_svhn.final = nn.Linear(4096, nlabel)

  ## Update from scratch
  else:
    print("Learning from scratch")
    # create the network
    tag = "simpleCNN_svhn_scratch"
    net_svhn = SimpleCNN(n_out=nlabel, n_in=3)

  net_svhn.to(device)

  # Define learning components (to be used in the learning function)
  optimizer = torch.optim.Adam(net_svhn.parameters(), lr=1e-3)
  scheduler = lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5, verbose=True)

  # Learning
  max_epoch = 1

  learned_model, train_error, test_error, train_losses = \
  train_model_multiclass(net_svhn, nlabel,
                          trainloader, testloader,
                          optimizer, scheduler,
                          num_epochs=max_epoch, display=False)

  # Save last learned model
  torch.save(learned_model, os.path.join(img_path, 'network_{:s}.pth'.format(tag)))

```

  - Perte finale : 1.388943e+00
  - Temps d’entraînement : 35 secondes

- **LAST (Dernière couche)** :
  - Résultats proches de ceux obtenus avec FT après une seule époque, bien que l’ajustement se limite à la dernière couche.

- **SCRATCH (Depuis zéro)** :
  - Performances légèrement inférieures, montrant que le transfert d’apprentissage est plus efficace.

```python
## Computes the performance of the three learning strategies
for tag in ['simpleCNN_svhn_ft', 'simpleCNN_svhn_scratch', 'simpleCNN_svhn_last']:
  net_svhn = torch.load(os.path.join(img_path, 'network_{:s}.pth'.format(tag)))
  # Evaluation
  net_svhn.eval()

  cm = np.zeros((nlabel,nlabel),dtype=int)
  with torch.no_grad():  ### no need to compute gradients
      for inputs, targets in tqdm(testloader, ncols=80):
          inputs = inputs.to(device)
          outputs = net_svhn(inputs)
          _,pred = outputs.max(1)
          cm += confusion_matrix(pred.cpu().numpy(),targets.cpu().numpy(),labels = list(range(nlabel)))

  print(f"Accuracy of {tag} = {100*accuracy(cm):.1f}%")
```

Les performances des trois stratégies ont été évaluées sur l'ensemble de test en calculant la matrice de confusion et la précision globale. Les résultats sont résumés dans le tableau ci-dessous :

## Analyse des résultats

   - Cette stratégie offre la meilleure précision (79.7%), car elle permet une adaptation complète des paramètres du réseau à la nouvelle tâche.

3. **Dernière couche (LAST)** :
   - La précision de 73.1% est proche de celle obtenue à partir de zéro, ce qui indique que les caractéristiques apprises sur GTSRB ne sont pas totalement adaptées à SVHN sans ajustement global.

La comparaison montre que le **Fine-tuning** est la stratégie la plus performante, car elle permet une adaptation complète et efficace à la nouvelle tâche.
